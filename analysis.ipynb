Revenue by Region

```python
revenue_region = (
    df.groupby("region")["revenue"]
      .sum()
      .sort_values(ascending=False)
)

revenue_region
```

Monthly Revenue Trend

```python
df["visit_date"] = pd.to_datetime(df["visit_date"])

monthly_revenue = (
    df.groupby(df["visit_date"].dt.to_period("M"))
      ["revenue"]
      .sum()
)

monthly_revenue
```

Territory Contribution %

```python
territory = (
    df.groupby("region")["revenue"]
      .sum()
      .reset_index()
)

territory["Contribution_%"] = (
    territory["revenue"]
    /
    territory["revenue"].sum()
    *100
)

territory
```

Top Reps

```python
top_reps = (
    df.groupby("sales_rep_name")
      .agg({
          "revenue":"sum",
          "achieved_pct":"mean"
      })
      .sort_values(
          "revenue",
          ascending=False
      )
)

top_reps
```

Underperforming Reps

```python
top_reps[
    top_reps["achieved_pct"] < 180
]
```

Top Doctors

```python
doctor_analysis = (
    df.groupby("doctor_name")
      ["prescriptions"]
      .sum()
      .sort_values(
          ascending=False
      )
)

doctor_analysis
```

Doctor Segmentation

```python
doctor_df = (
    df.groupby("doctor_name")
      ["prescriptions"]
      .sum()
      .reset_index()
)

doctor_df["Segment"] = np.where(
    doctor_df["prescriptions"] > 2500,
    "High Value",
    np.where(
        doctor_df["prescriptions"] > 1500,
        "Medium Value",
        "Low Value"
    )
)

doctor_df
```

Product Revenue

```python
product_revenue = (
    df.groupby("medicine_name")
      ["revenue"]
      .sum()
      .sort_values(
          ascending=False
      )
)

product_revenue
```

ROI by Region

```python
roi_region = (
    df.groupby("region")
      .agg({
          "revenue":"sum",
          "marketing_spend":"sum"
      })
)

roi_region["roi"] = (
    roi_region["revenue"]
    /
    roi_region["marketing_spend"]
)

roi_region.sort_values(
    "roi",
    ascending=False
)
```

# Executive Summary & Business Recommendations

The Pharma Sales Force Effectiveness Analytics project was conducted to evaluate territory performance, sales representative effectiveness, physician contribution, product performance, and marketing ROI.

Analysis revealed that the South region generated the highest revenue (₹115.1M), indicating strong market penetration and commercial performance. However, the North region achieved the highest marketing ROI (10.22), demonstrating the most efficient utilization of marketing investments. This highlights the importance of balancing revenue growth with operational efficiency.

Physician analysis identified Roy Newman as the highest-value physician based on prescription volume, suggesting that a small group of physicians contributes significantly to overall business performance. Product analysis showed that HeartSafe is the leading revenue-generating medicine, making it a critical driver of portfolio growth.

Sales force analysis indicated strong overall performance, with all regions exceeding their assigned targets and East region achieving the highest average target attainment. This suggests effective field execution but also indicates that target-setting methodologies may require reassessment to ensure continued performance improvement.

### Key Recommendations

* Continue investing in high-revenue territories while replicating their successful strategies across other regions.
* Increase engagement with high-value physicians to maximize prescription generation and strengthen relationships.
* Optimize marketing investments by adopting best practices from high-ROI regions.
* Prioritize promotion and availability of top-performing products such as HeartSafe.
* Reassess sales targets and introduce stretch goals to drive continuous improvement in sales force performance.

This project demonstrates how pharmaceutical sales data can be transformed into actionable business insights that support territory optimization, physician targeting, sales force effectiveness, and strategic decision-making.
